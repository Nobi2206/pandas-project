```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_excel("paints_dataset_10col_5000rows.xlsx")
```

```python
df.isnull().sum()
```

```python
df['Price_INR'].fillna(df['Price_INR'].mean(), inplace=True)
```

```python
df['Volume_ml'].fillna(df['Volume_ml'].mean(), inplace=True)
```

```python
df['ShelfLife_months'].fillna(df['ShelfLife_months'].mean(), inplace=True)
```

```python
df['Brand'].fillna(df['Brand'].mode()[0], inplace=True)
```

```python
df['ProductType'].fillna(df['ProductType'].mode()[0], inplace=True)
```

```python
df['Finish'].fillna(df['Finish'].mode()[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.groupby('Brand')['Price_INR'].mean()
```

```python
df['ProductType'].value_counts()
```

```python
df[df['Price_INR'] == df['Price_INR'].max()]
```

```python
df.groupby('ProductType')['ShelfLife_months'].mean()
```

```python
df['Price_per_ml'] = df['Price_INR'] / df['Volume_ml']
df.head()
```

```python
df.to_csv("cleaned_paints_data.csv", index=False)
```
